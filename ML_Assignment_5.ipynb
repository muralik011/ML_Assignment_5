<b>2. Describe quantitative and qualitative data in depth. Make a distinction between the two.

Quantitative data refers to numeric data or  metrics which can be measured like height, weight, temperature, etc. Qualitative data refers to data that describe a characteristic like gender, satisfaction, etc.<br><br>
- We can apply aggregation measures like mean ,median on quantitative data, while proportion of each category can be calculated from qualitative data.
- We can perform arithmetic operations on quantitative data but not on qualitative data.
- We can use charts like histogram, box plot, scatter plot for quantitative data but we can use pie chart and bar chart for qualitative data

<b>4. What are the various causes of machine learning data issues? What are the ramifications?

- Noisy data due to outliers or errors in data- Makes it difficult for ML models to identify patterns while training
- Missing data - ML models cannot handle missing data so missing data should be imputed before training
- Multicollinearity - Few ML models especially linear models perform poorly if the data has multicollinearity
- Irrelevant features - Irrelevant features may caue overfitting. For example a model may learn from 'id' feature (which is irrelevant) and overfit and may not generalize well on unseen data.
- High dimensionality - Data with huge number of features may cause a model to overfit and also take up lot of time and space while training

<b>5. Demonstrate various approaches to categorical data exploration with appropriate examples.

```python
cat = pd.Categorical(['male', 'female', 'male', 'male', 'male'])
```

Exploring distribution

```python
dist = cat.value_counts()
print(dist)
```

bar chart

```python
from matplotlib import pyplot as plt
plt.bar(x=dist.index, height=dist)
plt.show()
```

pie chart

```python
plt.pie(dist,labels=dist.index)
plt.show()
```

<b>7. Describe the various methods for dealing with missing data values in depth.

- Imputing with mean,median, mode: When the values are missing completely at random and the missing vaues are less than 5% we can use this method.
- Imputing with a constant: If the values are not missing at random we can use a value like 999 or -1 or 'missing' to highlight the missing value
- Adding missing indicator: This method is followed in addition to mean/median imputation to highlight them in case the values are not missing at random.

<b>i. What is the IQR? What criteria are used to assess it?

IQR refers to inter quartile range. It is the difference between the third and the first quartile. IQR measures the spread across the central/middle portion of data

<b>ii. Describe the various components of a box plot in detail? When will the lower whisker
surpass the upper whisker in length? How can box plots be used to identify outliers?

- Minimum
- 1st quartile (25th percentile)
- 2nd quartile (50th percentile or median)
- 3rd quartile (75th percentile)
- Maximum<br><br>
Lower whisker surpasses the upper whisker in lenght when the data is left skewed.
<br><br>
Lower bound = Q1 - 1.5 * IQR<br>
Upper bound = Q3 + 1.5 * IQR<br>
The above formulae are used to identify outliers. Any value below lower bound and above upper bound are outliers.

<b>3. Use a cross-tab

Cross tab is used to visualize the frequency of items in multiple categories where one category forms the row and the other forms the column.

```python
df = pd.DataFrame({'gender':['male', 'female', 'male', 'female'], 'hair_color':['grey', 'grey', 'grey', 'black']})
pd.crosstab(index=df['gender'], columns=df['hair_color'])
```
